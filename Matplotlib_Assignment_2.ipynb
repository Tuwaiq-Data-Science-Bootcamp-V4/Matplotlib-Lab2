# Q2: Create a one line chart using two arrays. The first array represents x and the second array represents y
| x     |     y    |
| ----- | -------- |
| 1    | 1       |
| 2    | 2       |
| 3    | 3       |
| 4    | 5       |
| 5    | 8       |
| 6    | 13      |
| 7    | 20      |

```python
x=np.array([1,2,3,4,5,6,7])
y =np.array([1,2,3,5,8,13,20])

plt.plot(x, y, color='#99aa99', linestyle='-', linewidth=4)

plt.title('Line Chart', fontsize = 16)
plt.xlabel('x axis', fontsize = 14)
plt.ylabel('y axis', fontsize = 14)
plt.show()
```

# Q3: Create a two lines chart using the same previous arrays. 
But this time:
- Mark each point with a circle.
- Use dashed lines.
- Set the size of the marker (circle) to size 10.
- Set the width of the line to 3.
- Add legends. 'Normal' for array x and 'Fast' for array y.

```python
plt.plot(x, y, color = '#fa9999', linestyle = '--', marker = 'o', linewidth = 3, markersize = 10, label='Fast')
plt.plot(x, x, color = '#99aa99', linestyle = '--', marker = 'o', linewidth = 3, markersize = 10, label='Normal')

plt.title('Line Chart', fontsize = 20)
plt.xlabel('x axis', fontsize = 15)
plt.ylabel('y axis', fontsize = 15)
plt.legend()

plt.show()
```

# Q4: Display multiple line charts in one row using the charts given below.
- Give each chart a title 
- Add a super title 'my data visulaization assignment'. 

```python
#super title = my data visulaization assignment

#title = First Chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

#title = Second Chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

#title = Third Chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])

# Create a figure with subplots
f, ax = plt.subplots(1, 3, figsize=(10, 5))
#Fi
ax[0].plot(x_plot1, y_plot1, marker='^', color = '#99aa99')
ax[0].set_title('First Chart')
ax[0].set_xlabel('X axis')
ax[0].set_ylabel('Y axis')
#s
ax[1].plot(x_plot2, y_plot2, marker='o', color = '#fa9999')
ax[1].set_title('Second Chart')
ax[1].set_xlabel('X axis')
ax[1].set_ylabel('Y axis')
#t
ax[2].plot(x_plot3, y_plot3, marker='*', color = '#6cb3ff')
ax[2].set_title('Third Chart')
ax[2].set_xlabel('X axis')
ax[2].set_ylabel('Y axis')

f.suptitle('my data visulaization assignment', fontsize=20)
#better apearence
plt.tight_layout()
plt.show()
```

# Q6: Create a scatter plot  using two arrays. The first array represents carAge and the second array represents carSpeed, set a specific color for each dot by using an array of different colors and add labels for the plot
| carAge | carSpeed |
| ----- | -------- |
| 5    | 99       |
| 7    | 86       |
| 8    |87      |
| 7    | 88       |
| 2    | 111       |
| 17    | 86       |
| 2    | 103      |
| 9    | 87       |
| 4    | 94       |
| 11    | 78      |

- Set a new color for each dot by using an array of different colors.
- Change the size of the dots to size 350.
- Add labels for x-axis and y-axis.
- Adjust the transparency of the dots with the alpha argument. let the alpha value be 0.4

```python
carAge = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11])
carSpeed = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78])

colors = np.array(['#66a3ff','#fa9999','#99aa99','#98375F', '#2ef07c', '#ff1999','#6cb3ff','#9aff99','#ffcf99', '#cfcc39'])

plt.scatter(carAge, carSpeed, c = colors, s = 350, alpha = 0.4)

plt.title('Car Age and Speed')
plt.xlabel('Car Age')
plt.ylabel('Car Speed')

plt.show()
```

# Q7: Create a pie chart using two arrays. The first array represents python_libraries and the second array represents your skills out of 10 at each library
| python_libraries |   
| ------------ | 
| NumPy        | 
| Pandas       | 
| Matplotlib   | 
| Seaborn      | 
| Plotly       | 
| Scikit-learn | 

- Set a new color for each wedge by using an array of different colors.
- Pull your favorite library wedge 0.3 from the center of the pie.
- Add a shadow to the pie chart

```python
python_libraries = np.array(['NumPy', 'Pandas', 'Matplotlib', 'Seaborn', 'Plotlty', 'Scikit-learn'])
skills = np.array([8, 9, 7, 3, 2, 2])
colors = np.array(['#fa9999','#ffcf99', '#afcc39', '#6cb3ff','#99aa99','#fffa99'])

plt.pie(skills, labels = python_libraries, colors = colors, explode = [0, 0.3, 0, 0, 0, 0], shadow = True)
plt.title('My Skills in Different Python Libraries')

plt.show()
```
